# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school. 
---

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete_df = pd.DataFrame(school_data_complete)
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete_df["school_name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = len(school_data_complete_df["Student ID"].unique())
student_count
```

```python
# Calculate the total budget
the_total_budget = sum(school_data_complete_df["budget"].unique())
the_total_budget
```

```python
# Calculate the average (mean) math score
the_average_math_score = sum(school_data_complete_df["math_score"])/len(school_data_complete_df["Student ID"])
the_average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = sum(school_data_complete_df["reading_score"])/len(school_data_complete_df["Student ID"])
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = [{"Total Schools":school_count, 
                     "Total Students":student_count,
                     "Total Budget":the_total_budget,
                     "Average Math Score": the_average_math_score,
                     "Total Reading Score": average_reading_score,
                     "% Passing Math Score": passing_math_percentage,
                     "% Passing Reading Score": passing_reading_percentage,
                     "% Overall Passing": overall_passing_rate
                    }]

# Formatting
#district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
#district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary_df = pd.DataFrame(district_summary)
# Display the DataFrame
district_summary_df
```

## School Summary

```python
# Use the code provided to select the school type
school_type = pd.DataFrame(school_data.set_index(["school_name"])["type"])
school_type.rename(columns = {'type':'School type'}, inplace = True)
school_type.index.name = None
school_type
```

```python
# Calculate the total student count
total_students = pd.DataFrame(school_data_complete.groupby("school_name").count()["Student ID"])
total_students.rename(columns = {'Student ID':'Total students'}, inplace = True)
total_students.index.name = None
#total_students=pd.DataFrame(school_data_complete["school_name"].value_counts())
total_students
```

```python
# Calculate the total school budget and per capita spending
total_school_budget = pd.DataFrame(school_data_complete.groupby(["school_name"]).mean())[["budget"]]
total_school_budget.rename(columns = {'budget':'Total school budget'}, inplace = True)
per_student_budget = pd.DataFrame(total_school_budget["Total school budget"]/total_students["Total students"],columns=["Per student budget"])
per_student_budget.index.name = None
per_student_budget
```

```python
# Calculate the average test scores
average_math_score = pd.DataFrame(school_data_complete.groupby(["school_name"]).mean()["math_score"])
average_math_score.rename(columns = {'math_score':'Average math score'}, inplace = True)
average_math_score.index.name = None
average_math_score = pd.DataFrame(average_math_score["Average math score"].round(2))


average_reading_score = pd.DataFrame(school_data_complete.groupby(["school_name"]).mean()["reading_score"])
average_reading_score.rename(columns = {'reading_score':'Average reading score'}, inplace = True)
average_reading_score.index.name = None
average_reading_score = pd.DataFrame(average_reading_score["Average reading score"].round(2))
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_passing_math = school_data_complete[school_data_complete["math_score"] >= 70]
school_passing_math
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70]
school_passing_reading
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete.loc[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)]
passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
percentage_passing_math = pd.DataFrame(school_passing_math.groupby(["school_name"]).count()["student_name"] / total_students["Total students"] * 100,columns=["% passing math"])
percentage_passing_math.index.name = None
percentage_passing_math = pd.DataFrame(percentage_passing_math["% passing math"].round(2))

percentage_passing_reading = pd.DataFrame(school_passing_reading.groupby(["school_name"]).count()["student_name"] / total_students["Total students"] * 100,columns=["% passing reading"])
percentage_passing_reading.index.name = None
percentage_passing_reading = pd.DataFrame(percentage_passing_reading["% passing reading"].round(2))

percentage_overall_passing = pd.DataFrame(passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / total_students["Total students"] * 100,columns=["% overall passing"])
percentage_overall_passing.index.name = None
percentage_overall_passing = pd.DataFrame(percentage_overall_passing["% overall passing"].round(2))
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.concat([school_type,total_students,total_school_budget,per_student_budget,average_math_score,average_reading_score,percentage_passing_math,percentage_passing_reading, percentage_overall_passing], axis=1)
per_school_summary.sort_index(0,ascending=True)
# Display the DataFrame
per_school_summary
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary.sort_values(by=["% overall passing"],ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = per_school_summary.sort_values(by=["% overall passing"],ascending=True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
#Perform the necessary calculations to create a DataFrame that lists 
#the average math score for students of each grade level at each school.

# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by "school_name" and take the mean of each.
ninth_graders_scores = pd.DataFrame(ninth_graders.groupby(["school_name"]).mean()["math_score"])
tenth_graders_scores = pd.DataFrame(tenth_graders.groupby(["school_name"]).mean()["math_score"])
eleventh_graders_scores = pd.DataFrame(eleventh_graders.groupby(["school_name"]).mean()["math_score"])
twelfth_graders_scores = pd.DataFrame(twelfth_graders.groupby(["school_name"]).mean()["math_score"])

#Use the code to select only the `math_score`.
ninth_grader_math_scores = ninth_graders_scores["math_score"].round(2)
tenth_grader_math_scores = tenth_graders_scores["math_score"].round(2)
eleventh_grader_math_scores = eleventh_graders_scores["math_score"].round(2)
twelfth_grader_math_scores = twelfth_graders_scores["math_score"].round(2)

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.concat([ninth_grader_math_scores,tenth_grader_math_scores,eleventh_grader_math_scores,twelfth_grader_math_scores],axis=1,keys=["9th Grade Math Scores","10th Grade Math Scores","11th Grade Math Scores","12th Grade Math Scores"])

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade

```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per student budget"],bins=spending_bins,labels=spending_labels)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average math score"].round(2)
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average reading score"].round(2)
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% passing math"].round(2)
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% passing reading"].round(2)
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% overall passing"].round(2)
```

```python
# Assemble into DataFrame
spending_summary = pd.concat([spending_math_scores,spending_reading_scores,spending_passing_math,spending_passing_reading,overall_passing_spending],axis=1)


# Display results
spending_summary
```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
school_size_bins = per_school_summary.copy()
school_size_bins["School Size"] = pd.cut(school_size_bins["Total students"],bins=size_bins,labels=size_labels)
school_size_bins
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_size_bins.groupby(["School Size"]).mean()["Average math score"].round(2)
size_reading_scores = school_size_bins.groupby(["School Size"]).mean()["Average reading score"].round(2)
size_passing_math = school_size_bins.groupby(["School Size"]).mean()["% passing math"].round(2)
size_passing_reading = school_size_bins.groupby(["School Size"]).mean()["% passing reading"].round(2)
size_overall_passing = school_size_bins.groupby(["School Size"]).mean()["% overall passing"].round(2)
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.concat([size_math_scores,size_reading_scores,size_passing_math,size_passing_reading,size_overall_passing],axis=1)

# Display results
size_summary
```
